```python
import pandas as pd
import pickle
import os
import re
```

```python
ted_files = []

for element in os.listdir("Cluster_outputs"):
    if "ted_data-" in element:
        ted_files.append(element)
ted_files
```

```python
ted_dfs = []

for ted_file in ted_files:
    with open(f"Cluster_outputs/{ted_file}", "rb") as file:
        ted_data_file = pickle.load(file)
        ted_dfs.append(ted_data_file)
ted_dfs
```

```python
ted_dfs[2]["index"] = ted_dfs[2]["index"]+10
```

```python
concatened = pd.concat(ted_dfs)
concatened
```

```python
concatened.reset_index(inplace=True, drop=True)
```

```python
concatened.to_pickle("Cluster_outputs/ted_data.pickle")
```
